```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dag = []
for number in range(1,31):
    dag.append(number)
```

```python
A_start = 20
plus = 20
A = [20]
for x in range(1,30):
    bedrag = A_start + plus * x
    A.append(bedrag)
```

```python
B_start = 2
rente = 1.25
B = [2]
for x in range(1,30):
    bedrag_B = 2 * rente ** x
    B.append(bedrag_B)
```

```python
df=pd.DataFrame()
```

```python
df['dag'] = dag
df['A'] = A
df['B'] = B
```

```python
x = dag
y = A
z = B
plt.plot(x, y, label='A')
plt.plot(x,z, label='B')
plt.legend()
plt.annotate('600', (29, 650))
plt.annotate('1292', (27, 1240))
plt.title('Lineair vs Exponentieel')
plt.show()
```
